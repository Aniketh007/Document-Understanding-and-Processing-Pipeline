```python
import pytesseract
from PIL import Image
import io
import matplotlib.pyplot as plt
import PyPDF2
import re
import nltk
from pdf2image import convert_from_path
```

```python
def extract_text(filepath):
    txt = ""
    with open(filepath, 'rb') as file:
        pdf = PyPDF2.PdfReader(file)
        nums = len(pdf.pages)
        for num in range(nums):
            page = pdf.pages[num]
            txt += page.extract_text()
        print(txt)
    return txt
```

```python
def pdf_to_images(filepath):
    images = convert_from_path(filepath)
    return images

def ocr_from_images(images):
    text = ""
    for image in images:
        text += pytesseract.image_to_string(image)
    return text
```

```python
path="/kaggle/input/ocr-model/sample 3.pdf"
pdf=extract_text(path)
img=pdf_to_images(path)

images=ocr_from_images(img)
```

```python
txt=re.sub(r'\s',' ',images)
print(txt)
txt=re.sub(r'[^\x00-\x7F]+',' ',txt)
```

```python
images=images.replace('\n\n','\n')
print(images)
```

```python
if(len(pdf)>len(images)):
  data=pdf
else:
  data=images
```

```python
data=re.sub(r'[^\w\s\n]+', '',data)
print(data)
```

```python
nltk.download('punkt')
sentence=nltk.sent_tokenize(data)
print(sentence)
```

```python
tokenize_word=[nltk.word_tokenize(sent) for sent in sentence]
for i in tokenize_word:
  print(i)
```

```python
def llm_information_extraction(text, max_length=512):
    model = pipeline("text-generation", model="dslim/bert-base-NER")
    extracted_info = model(text, max_length=max_length, num_return_sequences=1)
    return extracted_info

extracted_text = llm_information_extraction(data)
```

```python
def llm_document_classification(text):
    classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
    category = classifier(text, candidate_labels=["category1", "category2", "category3"])
    return category
category=llm_document_classification(data)
```

```python
def llm_translation(text, source_language, target_language):
    translator = pipeline("translation", model=f"Helsinki-NLP/opus-mt-{source_language}-{target_language}")
    translated_text = translator(text)
    return translated_text
text=llm_translation(data, 'de', 'en')
```
